```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# creating a sample data set
data = {
    'Age' : [23, 25, 22, 30, 28, 35, 40, 23, 24, 22],
    'Height' : [165, 170, 160, 175, 180, 172, 168, 158, 165, 170],
    'weight'   : [65, 70, 60, 75,80, 72, 68, 58, 65, 70],
    'Gender': ['Male', 'Female', 'Female', 'Male', 'Male', 'Female', 'Male', 'Female', 'Female', 'Male'], 
    'Smoker' : ['No', 'Yes', 'No', 'No', 'Yes', 'No', 'Yes', 'No', 'Yes', 'N0']
}
```

```python
df = pd.DataFrame(data)
```

```python
# step 1
```

```python
print("Data Types:\n", df.dtypes)
print("\nSummary Statistics:\n", df.describe())
```

```python
# step 2
plt.figure(figsize = (10, 4))
```

```python
plt.subplot(1, 2, 1)
sns.histplot(df['Age'], kde = True, color = 'blue')
plt.title('Age Distribution (Discrete Data)')
```

```python
plt.subplot(1, 2, 2)
sns.histplot(df['Height'], kde = True, color = 'green')
plt.title('Height Distribution (Continuous Data)')
```

```python
plt.tight_layout()
plt.show()
```

```python
# setp 3
plt.figure(figure = (10, 4)) 
```

```python
plt.subplot(1, 2, 1)
sns.countplot(x = 'Gender', data = df, palette = 'Set1')
plt.title('Gender Count (Categorical Data)')
```

```python
plt.subplot(1, 2, 2)
sns.countplot(x = 'Smoker', data = df, palette = 'Set2')
plt.title('Smoker Count (Categorical Data)')
```

```python
plt.tight_layout()
plt.show()
```

```python
# step 4 
sns.scatterplot(x = 'Age', y = 'weight', hue = 'Gender', data = df)
plt.title('Scatter Plot of Age vs Weight')
plt.show()
```

```python
# step 5 
plt.figure(figure = (8, 6))
sns.scatterplot(x = 'Gender', y = 'Height',  data = df)
plt.title('Scatter Plot of Age vs Weight')
plt.show()
```
